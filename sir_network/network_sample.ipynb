```python
import numpy as np
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import ndlib.models.epidemics as ep
import ndlib.models.ModelConfig as mc
from bokeh.io import output_notebook, show
from ndlib.viz.bokeh.DiffusionTrend import DiffusionTrend
```

```python
g=nx.watts_strogatz_graph(10,3,1)
model = ep.SIRModel(g)
```

```python
cfg = mc.Configuration()
cfg.add_model_parameter('beta', 1) #感染率
cfg.add_model_parameter('gamma', 1) #回復率(2週間)
cfg.add_model_parameter("fraction_infected", 1/10) #初期感染率
model.set_initial_status(cfg)
```

```python
#ノードの位置固定
pos = {
        n: (np.cos(2*i*np.pi/12), np.sin(2*i*np.pi/12))
        for i, n in enumerate(g.nodes)
    }
```

```python
node_color=[]
for i in range(len(model.status)):
    if model.status[i]==0:
        node_color.append("lightblue")
    elif model.status[i]==1:
        node_color.append("orange")
    else:
        node_color.append("lightgreen")
nx.draw(g,node_color=node_color,pos=pos,with_labels = True)
plt.savefig("step0.png")
```
