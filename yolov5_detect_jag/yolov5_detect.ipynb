```python
#動画の再生
#ps="/content/yolov5/runs/detect/exp2/VID_20210111_085722.mp4"

#!ffmpeg -i "/content/yolov5/runs/detect/exp3/VID_20210111_085722.mp4" -vcodec vp9 "/content/yolov5/runs/detect/exp3/VID_20210111_085722.webm"

from IPython.display import HTML
import base64
import io

def play(file_path):
    video = io.open(file_path, 'r+b').read()
    encoded = base64.b64encode(video)
    return(HTML(data='''<video width="640" height="480" controls><source src="data:video/mp4;base64,{0}" type="video/mp4" /></video>'''.format(encoded.decode('ascii'))))

play('/content/yolov5/runs/detect/exp/test2.webm')
```
